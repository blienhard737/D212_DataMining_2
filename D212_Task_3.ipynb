```python
# Import all modules and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from csv
raw = pd.read_csv('C:/Users/blien/Documents/WGU/D212/Task3/medical_market_basket.csv',
         skiprows=lambda x: x % 2) 
raw.iloc[[28]].T
```

```python
raw.head()
```

```python
print(len(raw))
print(raw.shape)
```

```python
#converting dataframe into list
list1 = len(raw)
listfinal = raw.shape[1]
prescriptions = []
for i in range(0,list1):
    prescriptions.append([str(raw.values[i,j]) 
        for j in range(0,listfinal)])
tranencode = TransactionEncoder()
array = tranencode.fit_transform(prescriptions)
```

```python
print(tranencode.columns_)
print(len(tranencode.columns_))
```

```python
array.shape
```

```python
# drop nan value
df = pd.DataFrame(array, 
        columns=tranencode.columns_)
df = df.drop(['nan'],axis=1) 
print(df.info())
print(df.shape)
```

```python
df.head()
```

```python
# export clean data ready for market basket analysis
df.to_csv('C:/Users/blien/Documents/WGU/D212/Task3/medical_market_basket_final.csv',header=True)
```

```python
# Minimum support 3/1000 - minimum threshold 5/10 (sorted by lift)
prescript = apriori(df, min_support=3/1000, use_colnames=True)
rules = association_rules(prescript, metric='confidence', min_threshold=5/10)
rules.sort_values(by=['lift'], ascending=False).head(10)
```

```python
# Minimum support 3/100 - minimum threshold 1/100
prescript = apriori(df, min_support=3/100, use_colnames=True)
rules = association_rules(prescript, metric='confidence', 
                min_threshold=1/100)
rules.sort_values(by=['support'], ascending=False).head(10)
```

```python
# Minimum support 3/100 - minimum threshold 1/100 (sorted by confindence)
# top 3 rules sorted by lift
prescript = apriori(df, min_support=3/1000, use_colnames=True)
rules = association_rules(prescript, metric='confidence', 
                min_threshold=1/100)
rules.sort_values(by=['confidence'], ascending=False).head(10)
```

```python
# top 3 rules sorted by lift
prescript = apriori(df, min_support=3/1000, use_colnames=True)
rules = association_rules(prescript, metric='confidence', min_threshold=5/10)
rules.sort_values(by=['lift'], ascending=False).head(3)
```
